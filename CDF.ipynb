```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate x values (range of values)
x = np.linspace(-1, 1, 100)

# Compute the CDF values
cdf = norm.cdf(x, loc=mean, scale=std_dev)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(x, cdf, label='CDF', color='blue')
plt.title('Cumulative Distribution Function (CDF)\nNormal Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.grid(True)
plt.legend()
plt.show()
```
