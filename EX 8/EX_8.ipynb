```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, boxcox
```

```python
# ============================
# Step 1: Load dataset
# ============================
file_path = "/home/mtech_fde/Sarvesh_Projects/EX 8/diabetes.csv"
data = pd.read_csv(file_path)
```

```python
# ============================
# Step 2: Pearson's skewness formula
# ============================
pearson_skewness = {}

for col in data.columns:
    if np.issubdtype(data[col].dtype, np.number):
        mean = np.mean(data[col])
        median = np.median(data[col])
        std_dev = np.std(data[col], ddof=1)
        pearson_val = (3 * (mean - median)) / std_dev if std_dev != 0 else np.nan
        pearson_skewness[col] = pearson_val

pearson_df = pd.DataFrame.from_dict(pearson_skewness, orient="index", columns=["Pearson’s Skewness"])
print("\n=== Pearson’s Skewness ===\n")
print(pearson_df)
```

```python

# Box-Cox Transformation
boxcox_data = {}
lambdas = {}

for col in data.select_dtypes(include=[np.number]).columns:
    col_data = data[col].copy()
    
    # Box-Cox requires strictly positive values
    if (col_data <= 0).any():
        col_data = col_data + abs(col_data.min()) + 1  # shift to positive
    
    transformed, fitted_lambda = boxcox(col_data)
    boxcox_data[col] = transformed
    lambdas[col] = fitted_lambda
    
    # Plot before vs after
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    sns.histplot(col_data, kde=True, ax=axes[0], color="blue", bins=30)
    axes[0].set_title(f"Before Box-Cox: {col}")
    
    sns.histplot(transformed, kde=True, ax=axes[1], color="green", bins=30)
    axes[1].set_title(f"After Box-Cox: {col}\n(Lambda={fitted_lambda:.2f})")
    
    plt.tight_layout()
    plt.show()

# Convert Box-Cox lambdas to DataFrame
lambda_df = pd.DataFrame.from_dict(lambdas, orient="index", columns=["Box-Cox Lambda"])
print("\n=== Box-Cox Lambda Values ===\n")
lambda_df
```
