```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 100      # number of coin tosses
p = 0.5      # probability of heads

# X-axis values: number of heads from 0 to 100
x = np.arange(0, n + 1)

# PMF: P(X = k)
pmf_values = binom.pmf(x, n, p)

# CDF: P(X <= k)
cdf_values = binom.cdf(x, n, p)

# Plot PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(x, pmf_values, basefmt=" ")
plt.title('PMF of 100 Coin Tosses')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, 'b-')
plt.title('CDF of 100 Coin Tosses')
plt.xlabel('Number of Heads')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```
