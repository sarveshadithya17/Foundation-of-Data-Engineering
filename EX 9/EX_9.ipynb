```python

# Descriptive Statistics on diabetes.csv


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load dataset
data = pd.read_csv("diabetes.csv")
```

```python
# 1. Distribution of each column
print("\n===== Distribution Type Detection =====")
for col in data.columns:
    plt.figure(figsize=(7,4))
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
    
    # Shapiro-Wilk normality test
    stat, p = stats.shapiro(data[col])
    skewness = stats.skew(data[col])
    
    if p > 0.05:
        dist_type = "Normal Distribution"
    else:
        if skewness > 0.5:
            dist_type = "Right-Skewed Distribution"
        elif skewness < -0.5:
            dist_type = "Left-Skewed Distribution"
        else:
            dist_type = "Approximately Symmetric (Non-normal)"
    
    print(f"Column: {col}")
    print(f"  Shapiro-Wilk Test p-value: {p:.4f}")
    print(f"  Skewness: {skewness:.4f}")
    print(f"  => Distribution Type: {dist_type}")
    print("-"*60)
```

```python
# 2. Central Tendency (Mean, Median, Mode)
print("\n===== Central Tendencies =====")
for col in data.columns:
    mean = data[col].mean()
    median = data[col].median()
    mode = data[col].mode()[0]
    print(f"{col}: Mean={mean:.4f}, Median={median:.4f}, Mode={mode}")
```

```python
# 3. Measures of Variability
print("\n===== Measures of Variability =====")
for col in data.columns:
    variance = data[col].var()
    std_dev = data[col].std()
    coeff_var = (std_dev / data[col].mean()) * 100
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    skewness = stats.skew(data[col])
    
    print(f"{col}: Variance={variance:.4f}, CoeffVar={coeff_var:.2f}%, IQR={iqr:.4f}, Skewness={skewness:.4f}")
    
    # Boxplot
    plt.figure(figsize=(6,3))
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# 4. Kurtosis (with classification)
print("\n===== Kurtosis Analysis =====")
for col in data.columns:
    kurt = stats.kurtosis(data[col])  # Fishers definition (0 = normal)
    if kurt > 0:
        k_type = "Leptokurtic (heavy tails)"
    elif kurt < 0:
        k_type = "Platykurtic (light tails)"
    else:
        k_type = "Mesokurtic (normal-like)"
    
    print(f"{col}: Kurtosis={kurt:.4f} -> {k_type}")
    
    # Plot distribution with KDE for kurtosis visualization
    plt.figure(figsize=(7,4))
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f"Kurtosis Check - {col} ({k_type})")
    plt.xlabel(col)
    plt.show()
```
