Create Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = {
    'Math_Score': np.random.randint(50, 100, 25),
    'English_Score': np.random.randint(45, 95, 25),
    'Science_Score': np.random.randint(40, 100, 25),
    'Attendance (%)': np.random.randint(70, 100, 25),
    'Study_Hours_per_Week': np.random.randint(5, 25, 25)
}
df = pd.DataFrame(data)
print("Original Student Dataset:\n")
print(df)
```

Data Normalization

```python
normalized_df = df.copy()
for column in normalized_df.columns:
    mean = normalized_df[column].mean()
    std = normalized_df[column].std()
    normalized_df[column] = (normalized_df[column] - mean) / std
print("\nNormalized Student Dataset:\n")
print(normalized_df)
```

Show range of each feature

```python
print("\nRange of Each Normalized Feature:\n")
for column in normalized_df.columns:
    print(f"{column}: min = {normalized_df[column].min():.2f}, max = {normalized_df[column].max():.2f}")
```

Visualize the normalized data

```python
sns.set(style="whitegrid")
sns.pairplot(normalized_df)
plt.suptitle("Pairplot of Normalized Student Features", y=1.02)
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(normalized_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation (Normalized Data)")
plt.show()
```

Save normalized data

```python
normalized_df.to_csv("normalized_student_data.csv", index=False)
print("\nNormalized dataset saved as 'normalized_student_data.csv'")
```
